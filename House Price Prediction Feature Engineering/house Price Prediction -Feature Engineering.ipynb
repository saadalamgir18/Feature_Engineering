# Missing Values

```python
features_nan = [feature for feature in df.columns if df[feature].isnull().sum()>1 and df[feature].dtype == 'O']
for feature in features_nan:
    print(f"{feature}: {np.round(df[feature].isnull().mean(), 4)}% missing values.")
```

```python
def replace_cat_feature(df, features_nan):
    data = df.copy()
    data[features_nan] = data[features_nan].fillna('Missing')
    return data
dataset = replace_cat_feature(df, features_nan)
dataset[features_nan].isnull().sum()
```

```python
dataset.head()
```

```python
features_num_nan = [feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtype != 'O']
for feature in features_num_nan:
    print(f"{feature}: {np.round(dataset[feature].isnull().mean(), 4)}% missing values.")
```

```python
for feature in features_num_nan:
    median_val = dataset[feature].median()
    dataset[feature+'nan'] = np.where(dataset[feature].isnull(), 1,0)
    dataset[feature].fillna(median_val, inplace=True)
dataset[features_num_nan].isnull().sum()
```

```python
dataset.head(20)
```

# Temporal Features

```python
for feature in ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']:
    dataset[feature] = dataset['YrSold'] - dataset[feature]
```

```python
dataset.head()
```

```python
dataset[['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']].head()
```

# Numerical Features

```python
num_features = ['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']
for feature in num_features:
    dataset[feature] = np.log(dataset[feature])
```

```python
dataset.head()
```

```python
dataset[num_features].head()
```

# Handling Rare Categorical Feature

```python
cat_features = [feature for feature in dataset.columns if dataset[feature].dtypes == 'O']
cat_features
```

```python
for feature in cat_features:
    temp = dataset.groupby(feature)['SalePrice'].count()/len(dataset)
    temp_df = temp[temp>0.01].index
    dataset[feature] = np.where(dataset[feature].isin(temp_df), dataset[feature], 'rare_var')
```

```python
dataset.head(100)
```

```python
for feature in cat_features:
    labels_orderd = dataset.groupby([feature])['SalePrice'].mean().sort_values().index
    print(labels_orderd)
    labels_orderd = {k:i for i,k in enumerate(labels_orderd, 0)}
    print(labels_orderd)
    dataset[feature] = dataset[feature].map(labels_orderd)
```

```python
dataset.head(10)
```

# Feature Scaling

```python
from sklearn.preprocessing import MinMaxScaler
feature_scal = [feature for feature in dataset.columns if feature not in ['Id', 'SalePrice'] ]
scaler = MinMaxScaler()
print(feature_scal)
scaler.fit(dataset[feature_scal])
```

```python
len(feature_scal)
```

```python
data = pd.concat([dataset[['Id', 'SalePrice']].reset_index(drop=True), pd.DataFrame(scaler.transform(dataset[feature_scal]), columns=feature_scal)],
                axis = 1)
```

```python
data.head()
```

```python
len(data.columns)
```

# Feature Selection

```python
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel
```

```python
y_train = data[['SalePrice']]
```

```python
y_train
```

```python
x_train = data.drop(["Id", "SalePrice"], axis=1)
```

```python
x_train.head()
```

```python
feature_sel_model = SelectFromModel(Lasso(alpha=0.005, random_state=0))
feature_sel_model.fit(x_train, y_train)
```

```python
feature_sel_model.get_support()
```

```python
feature_sel_model.get_support()
```

```python
selected_feature = x_train.columns[(feature_sel_model.get_support())]
```

```python
len(selected_feature)
```

```python
print(f"Total features: {x_train.shape[1]}")
print(f"Selected Features: {len(selected_feature)}")
print(f"Features with coefficient shrink to 0: {np.sum(feature_sel_model.estimator_.coef_ == 0)}")
```
